## EASYOCR

```python
import easyocr
```

```python
def easyocrs(image):
    read=easyocr.Reader(['en'],gpu=True)
    res=read.readtext(image)
    img=cv2.imread(image)
    for detection in res:
        tl=tuple([int(val) for val in detection[0][0]])
        br=tuple([int(val) for val in detection[0][2]])
        text=detection[1]
        img=cv2.rectangle(img,tl,br,(0,255,0),3)
        img=cv2.putText(img,text,tl,font,.5,(0,0,0),1,cv2.LINE_AA)
    plt.figure(figsize=(5,5))
    plt.imshow(img)
    plt.show()
    return res
    
```
